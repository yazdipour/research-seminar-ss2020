```python
from multiprocessing import Process
import sys
import os
import pandas as pd
import numpy as np
import re
from IPython.display import clear_output
```

```python
dest_path='/data/shya6478/06/'
```

```python
df_rpl = None
def add_rpl(t,col):
    global df_rpl
    if df_rpl is None:
        df_rpl = pd.DataFrame(columns=col)
    df_rpl.append(t, ignore_index=False)

def save_rpl(filename):
    global df_rpl
    global dest_path
    if df_rpl is None:
        print('df_rpl is empty')
        return
    path = f'{dest_path}/{filename}_rpl_.csv'
    df_rpl.to_csv(path)
    print(f'done saving {path}')
```

```python
import os
def save_df(df, chunk, filename):
    global dest_path
    if not os.path.exists(dest_path):
        os.makedirs(dest_path)
    path = f'{dest_path}{filename}_{chunk}.csv'
    df.to_csv(path)
    print('>> SAVED: ' + path)
```

```python
is_covid_text = lambda t: re.search('corona|covid', t, re.IGNORECASE) is not None # corona\d*|covid\d*
```

```python
def skip_if_file_exist(chunk, filename):
    global dest_path
    if not os.path.exists(dest_path):
        return False
    path = f'{dest_path}{filename}_{chunk}.csv'
    return os.path.isfile(path)
```

```python
def handle_df(df, chunk_counter, filename):
        for index, t in df.iterrows():
            if is_covid_text(t.text) is False:
                df.drop(index, inplace=True)
                if np.isnan(t.in_reply_to_status_id) is False:
                    add_rpl(t, df.columns.values)
        save_df(df, chunk_counter, filename)
```

```python
from threading import Thread
def read_data(filename):
    chunk_counter = 0
    path='/data/twitter/ss2020/all/'
    for df in pd.read_json(path+filename, chunksize=10000,lines=True):
        chunk_counter = chunk_counter + 1
        if skip_if_file_exist(chunk_counter, filename): 
            print(f'skipped {filename}--chunk#{chunk_counter}')
            continue
        print(f'start {filename}--chunk#{chunk_counter}')
        Thread(target=handle_df, args=(df, chunk, filename, )).start()
    save_rpl(filename)
```

```python
jobs = []
for i in range(11): # 2 till 11
    filename=f'data{i+1}.json'
    print('starting p:'+filename)
    p = Process(target=read_data, args=(filename,))
    jobs.append(p)
    p.start()
```
